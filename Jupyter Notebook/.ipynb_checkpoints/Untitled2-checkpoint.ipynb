```python
import pandas as pd

daten = pd.read_csv('../PyCharm/daten.csv')

daten.describe()
```

```python
import random

genre1 = daten.loc[daten['genre'] == 1]
genre2 = daten.loc[daten['genre'] == 2]
genre3 = daten.loc[daten['genre'] == 3]
genre4 = daten.loc[daten['genre'] == 4]
genre5 = daten.loc[daten['genre'] == 5]

anzahl= [len(genre1), len(genre2), len(genre3), len(genre4), len(genre5)]

ueberschuss = len(genre1) - min(anzahl)

drops = []
k = 0

for i in range (ueberschuss - 1):
    index = random.randrange(len(genre1))
    drops.append(index)
    k += 1

daten_g = daten.drop(drops)
len(daten_g)
```
